```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "Age (Years)": [10, 11],
    "Planned (m)": [3.5, 4.8],
    "Actual (m)": [2.4, 2.8],
}

df = pd.DataFrame(data)


x = df["Age (Years)"]
y = df["Actual (m)"]
coefficients = np.polyfit(x, y, 1) 
slope, intercept = coefficients
state_record = 8.72

age_months = np.arange(12, 16 + 1/12, 1/12) 
age_years = np.arange(12,17,1)
actual_yearly = np.arange(12,17)

planned_distances = slope * age_months + intercept 
actual_distances = planned_distances + np.random.uniform(-0.1, 0.3, len(age_months)) 


projection_df = pd.DataFrame({
    "Age (Years)": age_months,
    "State_Record":state_record,
    "Actual_Distance":actual_distances,
    "Planned_Distances":planned_distances
})


plt.figure(figsize=(12, 6))
plt.scatter(df["Age (Years)"], df["Actual (m)"], color="purple", label="Historical Data")

plt.plot(age_years, actual_yearly, label="Actual (Year-wise)", marker='o', color="red")

plt.plot(age_months, actual_distances, label="Actual (Month-wise)", color="green")
plt.plot(age_months, planned_distances, label="Planned (Month-wise)", linestyle="--", color="blue")

plt.title("Long Jump Performance Projection (Age 12 to 16)")
plt.xlabel("Age (Years)")
plt.ylabel("Long Jump Distance (meters)")
plt.legend()
plt.grid()
plt.show()
```
